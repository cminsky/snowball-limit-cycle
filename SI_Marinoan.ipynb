```python
def plot_temps(results, Marinoan=(56+22,56+22+4), shade_alpha=0.1):
    fig, ax = plt.subplots(figsize=(8, 5))
    t, T = results['t'], results['T']
    ax.plot(t, T, color='k', alpha=1)
    
    # imported annotations
    Ti, To = results['Ti'][0], results['To'][0]
    annotate_temp(ax,t_max=np.max(t),Sturtian=True,Ti=Ti,To=To)
    
    # Add shaded snowball windows
    ax.axvspan(*(0,56), alpha=shade_alpha, color='k')
    ax.axvspan(*Marinoan, alpha=shade_alpha, color='k')
    
    # Annotate Marinoan
    y = 316  # vertical position of label
    tick_height = 2
    start, end = Marinoan[0], Marinoan[1]
    ax.hlines(y, *Marinoan, color='k')
    ax.plot([Marinoan[0], Marinoan[0]], [y - tick_height, y + tick_height], color='k')
    ax.plot([Marinoan[1], Marinoan[1]], [y - tick_height, y + tick_height], color='k')
    ax.text(np.mean(Marinoan), y + 3, "Marinoan", ha='center', va='bottom', fontsize=10)
    
    plt.show()
    return fig,ax
```

```python
# long time to see stabilization

params = {'W_LIP':10,'tau':22.7,'n':0,'n_LIP':0,'dt':1e4/1e6}
results = run_model(t_max=200,**params)

t = results['t']
snowball = results['snowball']
snowball_starts, snowball_ends, snowball_durations, interglacial_durations = get_times(t, snowball)

almost_end_time = snowball_ends[-2]
end_time = snowball_ends[-1]
last_interglacial = interglacial_durations[-1]

print(f"Second-to-last snowball ends at {almost_end_time:0.1f} Myr")
print(f"Last snowball ends at {end_time:0.1f} Myr")
print(f"Last interglacial lasts {last_interglacial:0.1f} Myr")

fig, ax = plot_temps(results, shade_alpha=0.05)
```

### Minimize W_LIP, infinite tau

```python
def has_snowballs(W_LIP_val):
    params = {'W_LIP': W_LIP_val, 'tau': np.inf, 'n': 0.2, 'n_LIP': 0}
    results = run_model(t_max=10, **params)
    snowball = results['snowball']
    return np.any(snowball)

min_W = None
W_vals = np.linspace(0.3176, 0.3178, 10)
for W in W_vals:
    flag = has_snowballs(W)
    print(W,flag)
    if not min_W and flag:
        min_W = W
```
